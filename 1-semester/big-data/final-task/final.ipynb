```python
def get_plot_values() -> tuple[list, list]:
    res_x = []
    res_y = []
    cur_year = 2019
    for _, row in index.iterrows():
        for idx, value in enumerate(row):
            if idx == 0:
                cur_year = round(value)
                continue

            if not pd.isna(value):
                res_y.append(datetime.date(year=cur_year, month=idx, day=1))
                res_x.append(value)

    return res_x, res_y
```

```python
x, y = get_plot_values()
print(y)
print(x)
```

```python
plt.plot(y, x)
plt.title("График инфляции")
plt.xlabel('Время')
plt.ylabel('Инфляция')
plt.legend()
plt.show()
```

```python
from scipy.stats import pearsonr
```

```python
inflation_values, _ = get_plot_values()
```

```python
def calc_key(date: datetime.date) -> str:
    return f"{date.year}-{date.month}"

def get_ruble_values() -> list:
    res_x = []
    res_y = []
    date_set = set()
    for index in reversed(data.index):
        row = data.iloc[index]
        date = row["data"]
        if date.year == 2018:
            continue

        if calc_key(date) not in date_set:
            res_x.append(row["data"])
            res_y.append(row["curs"])
            date_set.add(calc_key(date))

    res_x.pop()
    res_x.pop()
    res_y.pop()
    res_y.pop()
    return res_x, res_y
```

```python
dates, values = get_ruble_values()
print(len(dates))
```

```python
correlation, _ = pearsonr(values, inflation_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("vgsales.csv")
cleaned_data = data.dropna().drop_duplicates()

# Настраиваем стиль для графиков
sns.set_theme(style="whitegrid")

# Этап 3: Визуализация продаж разных категорий видеоигр Nintendo в сравнении с конкурентами
# Фильтруем данные для Nintendo и конкурентов
nintendo_data = cleaned_data[cleaned_data['Publisher'] == 'Nintendo']
competitors_data = cleaned_data[cleaned_data['Publisher'] != 'Nintendo']

# Сводим данные по жанрам
nintendo_genre_sales = nintendo_data.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
competitors_genre_sales = competitors_data.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

# Создаем график для Nintendo
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.barplot(y=nintendo_genre_sales.index, x=nintendo_genre_sales.values, palette="Blues_d")
plt.title('Nintendo Sales by Genre')
plt.xlabel('Global Sales (Millions)')

# Создаем график для конкурентов
plt.subplot(1, 2, 2)
sns.barplot(y=competitors_genre_sales.index, x=competitors_genre_sales.values, palette="Reds_d")
plt.title('Competitors Sales by Genre')
plt.xlabel('Global Sales (Millions)')

# Отображаем графики
plt.tight_layout()
plt.show()
```

```python
# Этап 4: Визуализация продаж видеоигр Nintendo в сравнении с конкурентами по годам

# Сводим данные по годам
nintendo_yearly_sales = nintendo_data.groupby('Year')['Global_Sales'].sum()
competitors_yearly_sales = competitors_data.groupby('Year')['Global_Sales'].sum()

# Создаем график продаж по годам
plt.figure(figsize=(14, 7))

# Построение продаж Nintendo по годам
nintendo_line, = plt.plot(nintendo_yearly_sales.index, nintendo_yearly_sales.values, label='Nintendo', color='blue', marker='o')

# Построение продаж конкурентов по годам
competitors_line, = plt.plot(competitors_yearly_sales.index, competitors_yearly_sales.values, label='Competitors', color='red', marker='x')

# Настройка графика
plt.title('Yearly Global Sales: Nintendo vs Competitors')
plt.xlabel('Year')
plt.ylabel('Global Sales (Millions)')
plt.legend(handles=[nintendo_line, competitors_line])
plt.grid(True)

# Отображаем график
plt.show()
```

```python
import numpy as np

# Этап 5: Визуализация продаж видеоигр Nintendo в сравнении с конкурентами в разных регионах

# Сводим данные по регионам
nintendo_regional_sales = nintendo_data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
competitors_regional_sales = competitors_data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Создаем столбчатые графики продаж по регионам
fig, ax = plt.subplots(figsize=(14, 7))

# Настройка ширины столбцов
bar_width = 0.35
index = np.arange(len(nintendo_regional_sales))

# Построение графика для Nintendo
nintendo_bars = plt.bar(index, nintendo_regional_sales, bar_width, label='Nintendo', color='blue')

# Построение графика для конкурентов
competitors_bars = plt.bar(index + bar_width, competitors_regional_sales, bar_width, label='Competitors', color='red')

# Настройка графика
plt.xlabel('Region')
plt.ylabel('Sales (Millions)')
plt.title('Regional Sales: Nintendo vs Competitors')
plt.xticks(index + bar_width / 2, nintendo_regional_sales.index)
plt.legend()

# Отображение графика
plt.tight_layout()
plt.show()
```
