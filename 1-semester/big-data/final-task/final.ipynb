```python
def get_plot_values() -> tuple[list, list]:
    res_x = []
    res_y = []
    cur_year = 2019
    for _, row in index.iterrows():
        for idx, value in enumerate(row):
            if idx == 0:
                cur_year = round(value)
                continue

            if not pd.isna(value):
                res_y.append(datetime.date(year=cur_year, month=idx, day=1))
                res_x.append(value)

    return res_x, res_y
```

```python
def calc_key(date: datetime.date) -> str:
    return f"{date.year}-{date.month}"

def get_ruble_values() -> list:
    res_x = []
    res_y = []
    date_set = set()
    for index in reversed(data.index):
        row = data.iloc[index]
        date = row["data"]
        if date.year == 2018:
            continue

        if calc_key(date) not in date_set:
            res_x.append(row["data"])
            res_y.append(row["curs"])
            date_set.add(calc_key(date))

    res_x.pop()
    res_x.pop()
    res_y.pop()
    res_y.pop()
    return res_x, res_y
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Диапазон возможных курсов Евро
euro_rates = np.linspace(60, 80, 400)

# Фиксированная цена в форвардном контракте
fixed_rate = 70

# Расчет выигрыша/проигрыша для компании "Радуга" (длинная позиция)
# Если курс Евро выше фиксированной цены, "Радуга" выигрывает разницу между рыночным курсом и фиксированной ценой
# Если курс Евро ниже фиксированной цены, "Радуга" теряет разницу
raduga_profit = (euro_rates - fixed_rate) * 1e6  # 1 млн Евро

# Расчет выигрыша/проигрыша для компании "Пересвет" (короткая позиция)
# Обратная ситуация к "Радуга"
peresvet_profit = (fixed_rate - euro_rates) * 1e6

# Создание графиков
plt.figure(figsize=(10, 6))

# График для компании "Радуга"
plt.plot(euro_rates, raduga_profit, label='Радуга (Длинная позиция)', color='blue')

# График для компании "Пересвет"
plt.plot(euro_rates, peresvet_profit, label='Пересвет (Короткая позиция)', color='red')

# Настройка графика
plt.title('Выигрыш/Проигрыш от форвардного контракта')
plt.xlabel('Курс Евро (рублей за Евро)')
plt.ylabel('Выигрыш/Проигрыш (рубли)')
plt.axhline(0, color='black', lw=0.5)  # Ось X
plt.axvline(70, color='grey', linestyle='--')  # Фиксированный курс
plt.legend()
plt.grid(True)

# Показать график
plt.show()
```

```python
from scipy.optimize import minimize

# Данные для активов
r0 = 1
E_r1 = 2
E_r2 = 3
sigma_r1 = 2
sigma_r2 = 4
corr_r1_r2 = 1/2

# Функция для расчета стандартного отклонения портфеля
def portfolio_std(w, sigma_r1, sigma_r2, corr_r1_r2):
    return ((w[0] * sigma_r1) ** 2 + (w[1] * sigma_r2) ** 2 + 2 * w[0] * w[1] * sigma_r1 * sigma_r2 * corr_r1_r2) ** 0.5

# Функция для расчета ожидаемой доходности портфеля
def portfolio_return(w, E_r1, E_r2):
    return w[0] * E_r1 + w[1] * E_r2

# Целевая функция для минимизации (минимизация отрицательной доходности для максимизации положительной)
def objective(w):
    return -portfolio_return(w, E_r1, E_r2)

# Ограничения
constraints = (
    {'type': 'eq', 'fun': lambda w: w[0] + w[1] - 1},  # Сумма весов равна 1
    {'type': 'ineq', 'fun': lambda w: 3 - portfolio_std(w, sigma_r1, sigma_r2, corr_r1_r2)}  # Стандартное отклонение <= 3
)

# Начальное приближение (50% на каждый актив)
initial_w = [0.5, 0.5]

# Ограничение для весов, они не могут быть меньше 0 и больше 1
bounds = ((0, 1), (0, 1))

# Решение оптимизационной задачи
result = minimize(objective, initial_w, method='SLSQP', bounds=bounds, constraints=constraints)

# Если оптимизация была успешной, выводим результаты
if result.success:
    optimal_w = result.x
    optimal_std = portfolio_std(optimal_w, sigma_r1, sigma_r2, corr_r1_r2)
    optimal_return = -objective(optimal_w)  # Меняем знак обратно на положительный
else:
    optimal_w = None
    optimal_std = None
    optimal_return = None

optimal_w, optimal_std, optimal_return
```
