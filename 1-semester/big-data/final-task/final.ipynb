```python
index.rename(columns={
    " Январь": 1,
    " Февраль": 2,
    " Март": 3,
    " Апрель": 4,
    " Май": 5,
    " Июнь": 6,
    " Июль": 7,
    " Август": 8,
    " Сентябрь": 9,
    " Октябрь": 10,
    " Ноябрь": 11,
    " Декабрь": 12,
}, inplace=True, errors="raise")
```

```python
index.head()
```

```python
def get_plot_values() -> tuple[list, list]:
    res_x = []
    res_y = []
    cur_year = 2019
    for _, row in index.iterrows():
        for idx, value in enumerate(row):
            if idx == 0:
                cur_year = round(value)
                continue

            if not pd.isna(value):
                res_y.append(datetime.date(year=cur_year, month=idx, day=1))
                res_x.append(value)

    return res_x, res_y
```

```python
x, y = get_plot_values()
print(y)
print(x)
```

```python
plt.plot(y, x)
plt.xlabel('Время')
plt.ylabel('Инфляция')
plt.show()
```

```python
from scipy.stats import pearsonr
```

```python
inflation_values, _ = get_plot_values()
```

```python
def calc_key(date: datetime.date) -> str:
    return f"{date.year}-{date.month}"

def get_ruble_values() -> list:
    res_x = []
    res_y = []
    date_set = set()
    for index in reversed(data.index):
        row = data.iloc[index]
        date = row["data"]
        if date.year == 2018:
            continue

        if calc_key(date) not in date_set:
            res_x.append(row["data"])
            res_y.append(row["curs"])
            date_set.add(calc_key(date))

    res_x.pop()
    res_x.pop()
    res_y.pop()
    res_y.pop()
    return res_x, res_y
```

```python
dates, values = get_ruble_values()
print(len(dates)
```

```python
correlation, _ = pearsonr(values, inflation_values)
```
